#### 5. A function that given a userid and movieid, returns a list of userids for other users that rated the movie identified by the provided movieid with the same score.

```python
def ret_user_with_same_rating(userID, movieID):
    given_user_rating = data[(data.user_id == userID) & (data.movie_id == movieID)].rating.values[0]
    return data[(data.movie_id == movieID) & (data.rating == given_user_rating)]['user_id']
```

#### 6 Some other statistic, figure, aggregate, or plot that you created using this dataset,along with a short description of what interesting observations you derived from it.

 I was curious to see the comparision of ratings by men and women on the genres on movies i like i.e. Comdey and Horror. I present my visualiation below , after which i will summarize my analysis

```python
comedy_movie_genre = data[data.genres == 'Comedy']
```

```python
comedy_movie_genre.head()
```

```python
pivoted_comedy = comedy_movie_genre.pivot_table(index=['movie_id', 'title'],
                           columns=['gender'],
                           values='rating',
                           fill_value=0)
```

```python
pivoted_comedy.head()
```

```python
pivoted_comedy['diff'] = pivoted_comedy.M - pivoted_comedy.F
pivoted_comedy.head()
```

```python
pivoted_comedy.reset_index('movie_id', inplace=True)
```

```python
most_50_comedy = pivoted_comedy.groupby('movie_id').size().sort_values(ascending=False)[:50]
```

```python
disagreements = pivoted_comedy[pivoted_comedy.movie_id.isin(most_50_comedy.index)]['diff']
disagreements.sort_values().plot(kind='barh', figsize=[9, 15])
plt.title('Male vs. Female Avg. Ratings\n(Difference > 0 = Favored by Men)')
plt.ylabel('Title')
plt.xlabel('Average Rating Difference');
```

```python
horror_movie_genre = data[data.genres == 'Horror']
```

```python
pivoted_horror = horror_movie_genre.pivot_table(index=['movie_id', 'title'],
                           columns=['gender'],
                           values='rating',
                           fill_value=0)
```

```python
pivoted_horror['diff'] = pivoted_horror.M - pivoted_horror.F
pivoted_horror.head()
```

```python
pivoted_horror.reset_index('movie_id', inplace=True)
```

```python
most_50_horror = pivoted_horror.groupby('movie_id').size().sort_values(ascending=False)[:50]
```

```python
disagreements_horror = pivoted_horror[pivoted_horror.movie_id.isin(most_50_horror.index)]['diff']
disagreements_horror.sort_values().plot(kind='barh', figsize=[9, 15])
plt.title('Male vs. Female Avg. Ratings\n(Difference > 0 = Favored by Men)')
plt.ylabel('Title')
plt.xlabel('Average Rating Difference');
```
